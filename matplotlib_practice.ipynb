```python
%matplotlib inline
```

```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#Set the x_axis to a list of strings for each month

x_axis = ["Jan","Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot

plt.plot(x_axis,y_axis)
```

```python
#Create the plot with ax.plt()

#fig,ax = plt.subplots() unpacks the tuple into the variables fig and ax (fig references the figure and ax the axes
# of the figure and allows us to annotate the figure)
fig,ax = plt.subplots() #plt.subplots returns a tuple that contains a figure and axes object(s)

ax.plot(x_axis,y_axis) #plots the data from the x and y axes.
```

```python
#the code above can be also be written as follows:

fig = plt.figure()
ax = fig.add_subplot()

```

```python
#Create the plot with ax.plt()

ax = plt.axes()

ax.plot(x_axis,y_axis)
```

```python
#Create the plot, be careful with plt.show because it should be used only one time because it can improper your graph

plt.plot(x_axis,y_axis)
plt.show()
```

```python
#Create the plot.

plt.plot(x_axis, y_axis, marker = '*', color = 'b', linewidth = 2, label = 'Boston')

#Create labels for the x and y axes.

plt.xlabel('Date')
plt.ylabel('Fare($)')

#Set the y limit between 0 and 45.

plt.ylim(0,45)

#Create a title 

plt.title('PyBer Fare by Month')

#Add grid.

plt.grid()

#Add the legend

plt.legend()
```

```python
#Create the plot.

fig,ax = plt.subplots(figsize=(8,5)) #plt.subplots returns a tuple that contains a figure and axes object(s)

ax.plot(x_axis,y_axis, label = 'Boston', marker = "D", color = 'g') #plots the data from the x and y axes.

ax. set_ylim(0, 45)

ax.grid()

ax.set_xlabel('Date')

ax.set_ylabel('Fare ($)')

ax.set_title('Pyber Fare by Month')

ax.legend()
```

```python
#Doing Bar Charts

plt.bar(x_axis,y_axis)
```

```python
#Annotate in the bar chart

#Create the plot

plt.bar(x_axis,y_axis, color='g', label='Boston')

#Create labels for the x and y axes.

plt.xlabel('Date')
plt.ylabel('Fare($)')

#Create a title 

plt.title('PyBer Fare by Month')

#Add the legend

plt.legend()
```

```python
#Creating a horizontal bar chart

plt.barh(x_axis,y_axis)
```

```python
#Invert the y axis to get the Jan-Dec dates in order

#Create the plot

plt.barh(x_axis,y_axis, color = 'magenta', label = 'Boston', )

plt.gca().invert_yaxis() #the method gca() means 'get currents axes'

plt.ylabel('Date')
plt.xlabel('Fare($)')

#Create a title 

plt.title('PyBer Fare by Month')

#Add the legend

plt.legend()
```

```python
#Creating bar charts with the object oriented approach

fig,ax =plt.subplots()

ax.bar(x_axis,y_axis, color = 'k', label = 'Boston')

```

```python
#Create a horizontal Bar chart with object oriented

fig,ax = plt.subplots()

ax.barh(x_axis,y_axis)
```

```python
fig,ax = plt.subplots()

ax.barh(x_axis,y_axis,color = 'cyan', label = 'Chicago')

ax.set_ylabel('Date')

ax.set_xlabel('Fare ($)')

ax.set_title('Pyber Fare by Month')

ax.legend()

ax.invert_yaxis() 
```

```python
#Scatter plots

plt.plot(x_axis,y_axis, 'o') #or you can use plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(y_axis,x_axis, color = 'r', label = 'Chicago')

#Create labels for the x and y axes.

plt.xlabel('Date')
plt.ylabel('Fare($)')

#Create a title 

plt.title('PyBer Fare by Month')


#Add the legend

plt.legend()

#Invert the y axis

plt.gca().invert_yaxis()
```

```python
#Creating bubble charts

plt.scatter(x_axis, y_axis, s=y_axis)

```

```python
#Creating bubble charts with larger sizes

y_axis_larger = []

for data in y_axis:
    
    y_axis_larger.append(data*3)
    

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#W can do the above utilizing the reading comprehension used in python in which
#new_list = [expression for item in list if conditional]

plt.scatter(x_axis, y_axis, s = [i *3 for i in y_axis])
```

```python
#Create scatter plots using Object-Oriented interface

fig, ax = plt.subplots()

ax.scatter(x_axis,y_axis)

```

```python
#Create bubble charts using Object Oriented interface


fig, ax = plt.subplots()

ax.scatter(y_axis,x_axis, s = [i*5 for i in y_axis ], color = 'skyblue', alpha = 0.8, edgecolors = 'black', label = 'Boston')

ax.set_ylabel('Date')

ax.set_xlabel('Fare ($)')

ax.set_title('Pyber Fare by Month')

ax.legend()

ax.invert_yaxis() 

ax.set_xlim(0,50)
```

```python
#Create a pie chart

plt.subplots(figsize=(8,8))

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values=[0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]

plt.pie(y_axis, labels=x_axis, explode = explode_values, autopct = '%.1f%%', colors = colors)

plt.show()
```

```python
#Create a pie chart with object oriented method

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values=[0, 0, .3, 0, 0, 0, 0.3, 0, 0, 0, 0, 0]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,labels=x_axis, explode = explode_values, autopct = '%.1f%%', colors = colors, counterclock = False, shadow = True, startangle= 90)


plt.show()
```
